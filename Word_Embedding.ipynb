```python
from sklearn.feature_extraction.text import CountVectorizer
corpus=[

        "I love machine Learning",
        "machine Learning is Fun",
        "I love coding in python",
]
```

```python
vectorizer= CountVectorizer()
X= vectorizer.fit_transform(corpus)
print ("Vocabulary-", vectorizer.vocabulary_)

print("Bow matrix" , X.toarray())
```

### TF-IDF (Term Frequency-Inverse Document Frequency)

TF-IDF is another method for word embedding. It considers not only the frequency of a word in a document but also how important the word is across the entire corpus.

**Pros:**
- Captures the importance of words by considering their frequency within a document and rarity across all documents.
- Often performs better than BOW for tasks like information retrieval and text classification.

**Cons:**
- Still does not capture semantic meaning or context.
- Can be sensitive to stop words (common words like "the", "a", etc., although these can be handled).

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus=[

        "I love machine Learning",
        "machine Learning is Fun",
        "I love coding in python",
]

tf_idf_vector = TfidfVectorizer()
X1 =tf_idf_vector.fit_transform(corpus)
print(tf_idf_vector)
print (X1.toarray())

```

```python
import pandas as pd

feature_names = tf_idf_vector.get_feature_names_out()
df= pd.DataFrame(X1.toarray(), columns=feature_names)
df

```

```python
doc_index = 0
tfidf_scores = X1[doc_index].toarray().flatten()
print(tfidf_scores)
```

```python
import  nltk
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

print(stop_words)
```

```python
cleaned_corpus=[]

for sent in corpus:
  words = word_tokenize(sent.lower())
  filetred_words = [word for word in words if word not in stop_words]
  cleaned_corpus.append(' '.join(filetred_words))

print(cleaned_corpus)
```
