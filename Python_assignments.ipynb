```python
num = int(input("Enter a number: "))

if num < 2:
    print("Not prime")
else:
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            print("Not prime")
            break
    else:
        print("Prime")
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print(factorial(5))
```

```python
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))
```

```python
def list_sum(lst):
    return sum(lst)

print(list_sum([1, 2, 3, 4]))
```
